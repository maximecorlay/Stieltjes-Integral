```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour calculer S(n)
def S(n):
    somme = 0
    for k in range(1, n+1):
        terme = np.exp((2*k-1)/n) * (np.arctan(k/n) - np.arctan((k-1)/n))
        somme += terme
    return somme

# Listes pour stocker les valeurs de n et S(n)
n_values = range(10, 1001)
S_values = [S(n) for n in n_values]

# Affichage du graphe
plt.plot(n_values, S_values, label="S(n)")
plt.xlabel("n")
plt.ylabel("S(n)")
plt.title("S(n) en fonction de n")
plt.grid(True)
plt.legend()
plt.show()
```
